```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
import numpy as np
mean=np.mean(df['total_bill'])
std=np.std(df['total_bill'])
print(mean,std)
```

```python
normalized_data=[]
for i in list(df['total_bill']):
  z_score=(i-mean)/std
  normalized_data.append(z_score)
```

```python
normalized_data
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

```python
df.head()
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['total_bnill','tip'])
```

```python
scaler.transform([[13,14]])
```
